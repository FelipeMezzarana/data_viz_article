```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
df_pokemons = pd.read_csv('pokemon.csv')
df_pokemons.sample(3)
```

# Bar Plot

**Question to be answered:**
+ What types of Pokemon have the highest attack?

```python
df_pokemons_att = (df_pokemons.groupby('Type 1', as_index = False).Attack.mean())

df_pokemons_att.head(3)
```

**First Steap**

```python
sns.barplot(x = df_pokemons_att['Attack'],
            y = df_pokemons_att['Type 1'])

# Saving figure to disk
plt.savefig(r'img\first_bar_plot.png',facecolor= '#ffffff',
            edgecolor = '#ffffff',bbox_inches ='tight')
plt.show()
```

**Second Steap**

```python
# Ordering
df_pokemons_att2 = df_pokemons_att.copy()
df_pokemons_att2.sort_values(by = ['Attack'], ascending = False, inplace =True)
df_pokemons_att2.reset_index(drop = True, inplace = True) 
# Filtering top 10
df_pokemons_att_top10 = df_pokemons_att2.loc[:10,]

# Creating the new Bar Plot
# Declaring a figure can be very useful to elegantly arrange the elements of your chart
fig, ax1 = plt.subplots(figsize=(12,9))

sns.barplot(x = df_pokemons_att_top10['Attack'],
            y = df_pokemons_att_top10['Type 1'],
            color = '#b2b2d9', # This parameter can also be used to define more than one color
            ax = ax1 # A fig. can contain several axis (vizualizations)
           )

# An axis object has many attributes and methods that allow us to customize our views
ax1.set_title("Average Attack by Pokemon Type",fontsize = 22)
ax1.tick_params(labelsize=16)

# Saving figure to disk
plt.savefig(r'img\second_bar_plot.png',facecolor= '#ffffff',
            edgecolor = '#ffffff',bbox_inches ='tight')
plt.show()
```

**Third Steap**

```python
# Declaring a figure can be very useful to elegantly arrange the elements of your chart
fig, ax1 = plt.subplots(figsize=(12,9))

sns.barplot(x = df_pokemons_att_top10['Attack'],
            y = df_pokemons_att_top10['Type 1'],
            color = '#b2b2d9', # This parameter can also be used to define more than one color
            ax = ax1 # A fig. can contain several axis (vizualizations)
           )


# An axis object has many attributes and methods that allow us to customize our views
ax1.set_title("Average Attack by Pokemon Type",
              fontsize = 22,
              loc = 'left', # Move title position to left
              x=-0.01, # Adjusting the x position of the title to be aligned with xaxis
              pad = 10 # The offset of the title from the top of the Axes
             )
ax1.tick_params(labelsize=16)  
ax1.xaxis.tick_top() # Moving xaxis to top
ax1.set(xlabel=None,ylabel=None) # Removing x and y labels
sns.despine(bottom = True, left = True) # Removing all borders

# Saving figure to disk
plt.savefig(r'img\third_bar_plot.png',facecolor= '#ffffff',
            edgecolor = '#ffffff',bbox_inches ='tight')
plt.show()
```

**Before\After comparison**

```python
# Declaring a figure can be very useful to elegantly arrange the elements of your chart
fig, (ax0, ax1) = plt.subplots(1,2,figsize=(26,9))


sns.barplot(x = df_pokemons_att['Attack'],
            y = df_pokemons_att['Type 1'],
            ax = ax0)


sns.barplot(x = df_pokemons_att_top10['Attack'],
            y = df_pokemons_att_top10['Type 1'],
            color = '#b2b2d9', # This parameter can also be used to define more than one color
            ax = ax1 # A fig. can contain several axis (vizualizations)
           )


# An axis object has many attributes and methods that allow us to customize our views
ax1.set_title("Average Attack by Pokemon Type",
              fontsize = 22,
              loc = 'left', # Move title position to left
              x=-0.01, # Adjusting the x position of the title to be aligned with xaxis
              pad = 10 # The offset of the title from the top of the Axes
             )
ax1.tick_params(labelsize=16) # Changing 
ax1.xaxis.tick_top() # Moving xaxis to top
ax1.set(xlabel=None,ylabel=None) # Removing x and y labels
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_visible(False)

plt.text(-100, -3.5, "Before", size= 32, weight='bold')
plt.text(50, -3.5, "After", size= 32, weight='bold')

plt.savefig(r'img\before_after.png',facecolor= '#ffffff',edgecolor = '#ffffff',bbox_inches ='tight')
plt.show()
```
